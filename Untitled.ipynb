```python
def find_images_with_text(directory, text):
    """
    Parcourt récursivement un répertoire pour trouver des fichiers contenant un texte spécifique dans leur nom.

    Paramètres:
    directory: Le répertoire racine à parcourir.
    text: Le texte à rechercher dans les noms de fichiers.
    
    Retourne: 
    Une liste des chemins des fichiers dont le nom contient le texte spécifié.
    """
    image_list = []

    # Utilisation de os.walk pour parcourir tous les sous-dossiers et fichiers
    for root, dirs, files in os.walk(directory):
        for file in files:
            # Vérification si le texte spécifié est dans le nom du fichier
            if text in file:
                # Ajout du chemin complet du fichier à la liste
                image_list.append(os.path.join(root, file))

    return image_list
```

```python
def show_random_images_with_masks(df, n=1):
    """
    Affiche aléatoirement n images et leurs masques associés à partir d'un DataFrame.

    Paramètres:
    df: DataFrame contenant les colonnes 'image' et 'mask', avec les chemins des images.
    n: Nombre d'images à afficher (par défaut 1).
    """
    # Sélectionne n lignes aléatoires dans le DataFrame
    indices = random.sample(range(len(df)), n)
    selected_rows = df.iloc[indices]
    
    # Parcours des lignes sélectionnées pour afficher les images et les masques
    for _, row in selected_rows.iterrows():
        image_path = row['image']
        mask_path = row['mask']
        color_path = row['color']
        
        # Lecture de l'image et du masque
        image = cv2.imread(image_path)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)  # Charger le masque en niveaux de gris
        color = cv2.imread(color_path)
        
        # Affichage de l'image et du masque côte à côte
        plt.figure(figsize=(10, 5))
        
        # Affiche l'image
        plt.subplot(1, 3, 1)
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.title("Image")
        plt.axis("off")
        
        # Affiche le masque
        plt.subplot(1, 3, 2)
        plt.imshow(mask, cmap='gray')
        plt.title("Masque en gris")
        plt.axis("off")
        
        # Affiche le masque en couleurs
        plt.subplot(1, 3, 3)
        plt.imshow(color)
        plt.title("Masque en couleurs")
        plt.axis("off")
        
        # Affiche le graphique
        plt.show()
```

## Fonctions de métrique et de perte

```python
def dice_coeff(y_true, y_pred):
    smooth = 1.0
    y_true_f = K.flatten(K.cast(y_true, 'float32'))  # Conversion explicite en float32
    y_pred_f = K.flatten(K.cast(y_pred, 'float32'))  # Conversion explicite en float32
    intersection = K.sum(y_true_f * y_pred_f)
    score = (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)
    return score

def dice_loss(y_true, y_pred):
    return 1 - dice_coeff(y_true, y_pred)

def total_loss(y_true, y_pred):
    binary_crossentropy_loss = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    return binary_crossentropy_loss + (3 * dice_loss(y_true, y_pred))

def jaccard_score(y_true, y_pred):
    smooth = 1.0
    y_true_f = K.flatten(K.cast(y_true, 'float32'))
    y_pred_f = K.flatten(K.cast(y_pred, 'float32'))
    intersection = K.sum(y_true_f * y_pred_f)
    union = K.sum(y_true_f) + K.sum(y_pred_f) - intersection
    return (intersection + smooth) / (union + smooth)
```

```python
cats = {
    'void': [0, 1, 2, 3, 4, 5, 6],
    'flat': [7, 8, 9, 10],
    'construction': [11, 12, 13, 14, 15, 16],
    'object': [17, 18, 19, 20],
    'nature': [21, 22],
    'sky': [23],
    'human': [24, 25],
    'vehicle': [26, 27, 28, 29, 30, 31, 32, 33, -1]
}
```

```python
class seg_gen(Sequence):
    def __init__(self, dataframe, batch_size, img_height=256, img_width=512):
        self.dataframe = dataframe
        self.batch_size = batch_size
        self.img_height = img_height
        self.img_width = img_width

    def __len__(self):
        return int(np.ceil(len(self.dataframe) / float(self.batch_size)))

    def __getitem__(self, idx):
        batch = self.dataframe.sample(n=self.batch_size)
        batch_x, batch_y = [], []
        
        for _, row in batch.iterrows():
            img_path = row['image']
            mask_path = row['mask']
            
            # Chargement de l'image en 512x2048
            img = image.img_to_array(image.load_img(img_path, target_size=(self.img_height, self.img_width))) / 255.0
            
            # Chargement du masque en niveaux de gris et en 512x2048
            mask = image.load_img(mask_path, color_mode="grayscale", target_size=(self.img_height, self.img_width))
            mask = image.img_to_array(mask).squeeze()  # Assure qu'il s'agit bien d'un tableau 2D (512, 2048)

            # Initialisation du masque one-hot encodé
            mask_one_hot = np.zeros((self.img_height, self.img_width, 8))
            for j in range(-1, 34):
                if j in cats['void']:
                    mask_one_hot[:,:,0] = np.logical_or(mask_one_hot[:,:,0], (mask == j))
                elif j in cats['flat']:
                    mask_one_hot[:,:,1] = np.logical_or(mask_one_hot[:,:,1], (mask == j))
                elif j in cats['construction']:
                    mask_one_hot[:,:,2] = np.logical_or(mask_one_hot[:,:,2], (mask == j))
                elif j in cats['object']:
                    mask_one_hot[:,:,3] = np.logical_or(mask_one_hot[:,:,3], (mask == j))
                elif j in cats['nature']:
                    mask_one_hot[:,:,4] = np.logical_or(mask_one_hot[:,:,4], (mask == j))
                elif j in cats['sky']:
                    mask_one_hot[:,:,5] = np.logical_or(mask_one_hot[:,:,5], (mask == j))
                elif j in cats['human']:
                    mask_one_hot[:,:,6] = np.logical_or(mask_one_hot[:,:,6], (mask == j))
                elif j in cats['vehicle']:
                    mask_one_hot[:,:,7] = np.logical_or(mask_one_hot[:,:,7], (mask == j))
                    
            batch_x.append(img)
            batch_y.append(mask_one_hot.reshape(-1, 8))
        
        return np.array(batch_x), np.array(batch_y)
```

```python
class visualize(tf.keras.callbacks.Callback):
    def __init__(self, dataframe, img_height=256, img_width=512, orig_height=256, orig_width=512):
        super().__init__()
        self.dataframe = dataframe
        self.img_height = img_height
        self.img_width = img_width
        self.orig_height = orig_height
        self.orig_width = orig_width
        
        # Définir les 8 couleurs personnalisées pour chaque classe
        self.colors = [
            "#000000",  # Couleur pour la classe 0
            "#804080",  # Couleur pour la classe 1
            "#464646",  # Couleur pour la classe 2
            "#dcdc00",  # Couleur pour la classe 3
            "#6b8e23",  # Couleur pour la classe 4
            "#4682b4",  # Couleur pour la classe 5
            "#dc143c",  # Couleur pour la classe 6
            "#00008e"   # Couleur pour la classe 7
        ]

        # Noms des catégories
        self.category_names = [
            "Void", "Flat", "Construction", "Object",
            "Nature", "Sky", "Human", "Vehicle"
        ]
        
        # Créer une carte de couleurs personnalisée avec ListedColormap
        self.cmap = ListedColormap(self.colors)

    def calculate_iou_per_class(self, y_true, y_pred, num_classes=8):
        iou_scores = []
        for i in range(num_classes):
            y_true_class = (y_true == i)
            y_pred_class = (y_pred == i)
            
            intersection = np.logical_and(y_true_class, y_pred_class).sum()
            union = np.logical_or(y_true_class, y_pred_class).sum()
            
            # Calculer l'IoU pour la classe si l'union n'est pas vide
            iou = intersection / union if union > 0 else np.nan
            iou_scores.append(iou)
            
        # Remplacer les valeurs NaN par 0 pour les classes absentes
        iou_scores = [iou if not np.isnan(iou) else 0 for iou in iou_scores]
        return iou_scores

    def on_epoch_end(self, epoch, logs=None):
        print(f'\nGenerating output at epoch: {epoch}')
        
        # Sélection d'une image pour la visualisation
        sample_row = self.dataframe.sample(n=1, random_state=42).iloc[0]
        
        # Récupération des valeurs des colonnes 'image' et 'color' de cette ligne
        test_img_path = sample_row['image']
        color_img_path = sample_row['color']
        
        test_img = cv2.imread(test_img_path)
        color_img = cv2.imread(color_img_path)
        
        # Affichage de l'image et du masque côte à côte
        plt.figure(figsize=(20, 8))
        
        # Affiche l'image originale
        plt.subplot(2, 3, 1)
        plt.imshow(cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB))
        plt.title("Image d'origine")
        plt.axis("off")
        
        # Chargement et redimensionnement de l'image pour l'entraînement
        img = image.img_to_array(image.load_img(test_img_path, target_size=(self.img_height, self.img_width))) / 255.0
        pred_mask = unet.predict(np.expand_dims(img, axis=0))[0]  # Prédiction du masque
        
        # Extraction des classes prédominantes par pixel et redimensionnement
        y_pred = np.argmax(pred_mask, axis=-1)
        y_pred_resized = cv2.resize(y_pred, (self.orig_width, self.orig_height), interpolation=cv2.INTER_NEAREST)

        # Affichage du masque prédit avec la carte de couleurs personnalisée
        plt.subplot(2, 3, 2)
        plt.imshow(y_pred_resized, cmap=self.cmap)
        plt.title("Masque prédit")
        plt.axis("off")

        # Affiche le masque d'origine
        y_true = cv2.imread(color_img_path, cv2.IMREAD_GRAYSCALE)  # Chargement du masque d'origine en niveaux de gris
        y_true_resized = cv2.resize(y_true, (self.orig_width, self.orig_height), interpolation=cv2.INTER_NEAREST)
        plt.subplot(2, 3, 3)
        plt.imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB))
        plt.title("Masque d'origine")
        plt.axis("off")

        # Calcul des IoU par classe pour le masque prédit
        iou_scores_pred = self.calculate_iou_per_class(y_true_resized, y_pred_resized)

        # Calcul des IoU par classe pour le masque d'origine lui-même
        iou_scores_true = self.calculate_iou_per_class(y_true_resized, y_true_resized)

        # Affichage de l'histogramme des IoU pour le masque prédit
        plt.subplot(2, 3, 5)
        plt.bar(range(8), iou_scores_pred, color=self.colors)
        plt.title("IoU par classe (Prédit)")
        plt.xlabel("Classes")
        plt.ylabel("IoU")
        plt.xticks(range(8), self.category_names, rotation=45)  # Assure que les classes sont correctement étiquetées

        # Affichage de l'histogramme des IoU pour le masque d'origine
        plt.subplot(2, 3, 6)
        plt.bar(range(8), iou_scores_true, color=self.colors)
        plt.title("IoU par classe (Origine)")
        plt.xlabel("Classes")
        plt.ylabel("IoU")
        plt.xticks(range(8), self.category_names, rotation=45)

        plt.tight_layout()
        plt.show()
```

# Data generator avec VGG16Unet

```python
def VGG16Unet(input_shape=(256, 512, 3), num_classes=8):
    # Charger le modèle VGG16 sans la couche de classification
    vgg16 = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)
    
    # Extraire les couches de VGG16 pour créer l'architecture U-Net
    f1 = vgg16.get_layer("block1_pool").output  # Première couche de pooling
    f2 = vgg16.get_layer("block2_pool").output  # Deuxième couche de pooling
    f3 = vgg16.get_layer("block3_pool").output  # Troisième couche de pooling
    f4 = vgg16.get_layer("block4_pool").output  # Quatrième couche de pooling
    f5 = vgg16.get_layer("block5_pool").output  # Dernière couche de pooling
    
    # Décodeur U-Net
    up6 = UpSampling2D((2, 2))(f5)
    up6 = Conv2D(512, (3, 3), activation='relu', padding='same')(up6)
    merge6 = concatenate([up6, f4])
    conv6 = Conv2D(512, (3, 3), activation='relu', padding='same')(merge6)
    
    up7 = UpSampling2D((2, 2))(conv6)
    up7 = Conv2D(256, (3, 3), activation='relu', padding='same')(up7)
    merge7 = concatenate([up7, f3])
    conv7 = Conv2D(256, (3, 3), activation='relu', padding='same')(merge7)
    
    up8 = UpSampling2D((2, 2))(conv7)
    up8 = Conv2D(128, (3, 3), activation='relu', padding='same')(up8)
    merge8 = concatenate([up8, f2])
    conv8 = Conv2D(128, (3, 3), activation='relu', padding='same')(merge8)
    
    up9 = UpSampling2D((2, 2))(conv8)
    up9 = Conv2D(64, (3, 3), activation='relu', padding='same')(up9)
    merge9 = concatenate([up9, f1])
    conv9 = Conv2D(64, (3, 3), activation='relu', padding='same')(merge9)
    
    # Ajouter une dernière couche d'UpSampling pour obtenir la résolution finale de 256x512
    up10 = UpSampling2D((2, 2))(conv9)
    conv10 = Conv2D(num_classes, (1, 1), activation='softmax')(up10)
    
    # Construire le modèle
    model = Model(inputs=vgg16.input, outputs=conv10)
    return model
```

```python
class Visualize(tf.keras.callbacks.Callback):
    def __init__(self, dataframe, img_height=256, img_width=512, num_classes=8, orig_height=256, orig_width=512):
        super().__init__()
        self.dataframe = dataframe
        self.img_height = img_height
        self.img_width = img_width
        self.num_classes = num_classes
        self.orig_height = orig_height
        self.orig_width = orig_width
        
        # Définir les couleurs pour chaque classe
        self.colors = [
            "#1f77b4",  # Classe 0
            "#ff7f0e",  # Classe 1
            "#2ca02c",  # Classe 2
            "#d62728",  # Classe 3
            "#9467bd",  # Classe 4
            "#8c564b",  # Classe 5
            "#e377c2",  # Classe 6
            "#7f7f7f"   # Classe 7
        ]
        
        # Noms des catégories
        self.category_names = [
            "Void", "Flat", "Construction", "Object",
            "Nature", "Sky", "Human", "Vehicle"
        ]
        
        # Carte de couleurs personnalisée
        self.cmap = ListedColormap(self.colors)

    def calculate_iou_per_class(self, y_true, y_pred, num_classes=8):
        iou_scores = []
        for i in range(num_classes):
            y_true_class = (y_true == i)
            y_pred_class = (y_pred == i)
            
            intersection = np.logical_and(y_true_class, y_pred_class).sum()
            union = np.logical_or(y_true_class, y_pred_class).sum()
            
            # Calculer l'IoU pour la classe si l'union n'est pas vide
            iou = intersection / union if union > 0 else np.nan
            iou_scores.append(iou)
        
        # Remplacer les valeurs NaN par 0 pour les classes absentes
        iou_scores = [iou if not np.isnan(iou) else 0 for iou in iou_scores]
        return iou_scores

    def on_epoch_end(self, epoch, logs=None):
        print(f'\nGenerating output at epoch: {epoch}')
        
        # Sélectionner un exemple du dataframe pour la visualisation
        sample_row = self.dataframe.sample(n=1, random_state=42).iloc[0]
        test_img_path = sample_row['image']
        color_img_path = sample_row['color']
        
        test_img = cv2.imread(test_img_path)
        color_img = cv2.imread(color_img_path)
        
        # Préparer l'image pour la prédiction (redimensionnée pour le modèle)
        img = image.img_to_array(image.load_img(test_img_path, target_size=(self.img_height, self.img_width))) / 255.0
        pred_mask = self.model.predict(np.expand_dims(img, axis=0))[0]  # Prédiction du masque
        
        # Convertir la prédiction en classes dominantes par pixel
        y_pred = np.argmax(pred_mask, axis=-1)
        y_pred_resized = cv2.resize(y_pred, (self.orig_width, self.orig_height), interpolation=cv2.INTER_NEAREST)

        # Afficher les images et masques
        plt.figure(figsize=(15, 10))
        
        # Image originale
        plt.subplot(2, 3, 1)
        plt.imshow(cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB))
        plt.title("Image d'origine")
        plt.axis("off")
        
        # Masque prédit
        plt.subplot(2, 3, 2)
        plt.imshow(y_pred_resized, cmap=self.cmap)
        plt.title("Masque prédit")
        plt.axis("off")

        # Masque d'origine
        plt.subplot(2, 3, 3)
        plt.imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB))
        plt.title("Masque d'origine")
        plt.axis("off")

        # Calcul des IoU par classe pour le masque prédit et le masque d'origine
        y_true = cv2.imread(color_img_path, cv2.IMREAD_GRAYSCALE)
        y_true_resized = cv2.resize(y_true, (self.orig_width, self.orig_height), interpolation=cv2.INTER_NEAREST)
        iou_scores_pred = self.calculate_iou_per_class(y_true_resized, y_pred_resized)
        iou_scores_true = self.calculate_iou_per_class(y_true_resized, y_true_resized)

        # Graphique de l'IoU pour le masque prédit
        plt.subplot(2, 3, 5)
        plt.bar(range(self.num_classes), iou_scores_pred, color=self.colors)
        plt.title("IoU par classe (Prédit)")
        plt.xlabel("Classes")
        plt.ylabel("IoU")
        plt.xticks(range(self.num_classes), self.category_names, rotation=45)

        # Graphique de l'IoU pour le masque d'origine
        plt.subplot(2, 3, 6)
        plt.bar(range(self.num_classes), iou_scores_true, color=self.colors)
        plt.title("IoU par classe (Origine)")
        plt.xlabel("Classes")
        plt.ylabel("IoU")
        plt.xticks(range(self.num_classes), self.category_names, rotation=45)

        plt.tight_layout()
        plt.show()
```

```python
class DataGenerator(Sequence):
    def __init__(self, dataframe, batch_size=16, img_height=256, img_width=512, num_classes=8):
        self.dataframe = dataframe
        self.batch_size = batch_size
        self.img_height = img_height
        self.img_width = img_width
        self.num_classes = num_classes

    def __len__(self):
        return int(np.ceil(len(self.dataframe) / self.batch_size))

    def __getitem__(self, idx):
        batch = self.dataframe.sample(n=self.batch_size)
        batch_x, batch_y = [], []
        
        for _, row in batch.iterrows():
            img_path = row['image']
            mask_path = row['mask']
            
            # Charger et redimensionner l'image
            img = image.img_to_array(image.load_img(img_path, target_size=(self.img_height, self.img_width))) / 255.0
            
            # Charger et redimensionner le masque
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            mask = cv2.resize(mask, (self.img_width, self.img_height), interpolation=cv2.INTER_NEAREST)
            
            # One-hot encoding du masque
            mask_one_hot = np.zeros((self.img_height, self.img_width, self.num_classes), dtype=np.float32)
            for i in range(self.num_classes):
                mask_one_hot[:, :, i] = (mask == i).astype(np.float32)
            
            batch_x.append(img)
            batch_y.append(mask_one_hot)
        
        return np.array(batch_x), np.array(batch_y)
```

```python
# Initialisation du modèle
vgg16_unet = VGG16Unet(input_shape=(256, 512, 3), num_classes=8)
vgg16_unet.compile(optimizer='adam', loss='categorical_crossentropy', metrics=[dice_coeff, dice_loss, total_loss, jaccard_score, 'accuracy'])
```
